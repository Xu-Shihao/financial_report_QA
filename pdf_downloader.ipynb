```python
import re
import requests
from concurrent.futures import ThreadPoolExecutor

def find_arxiv_links(text):
    # 使用正则表达式匹配以https://arxiv.org开头，并以空格或换行符结尾的链接
    pattern = r'https://arxiv\.org\S*'
    links = re.findall(pattern, text)

    res_ = []
    for x in links:
        if x[-1] in [";", '"']:
            x = x[:-1]
        if x[-4:] == ".pdf":
            x = x[:-4]
        res_.append(x)
        
    res_ = set(res_)
    print("\n".join(res_))
    return res_

import requests

def download_pdf(arxiv_link):
    pdf_link = arxiv_link.replace('/abs/', '/pdf/')
    response = requests.get(pdf_link)
    if response.status_code == 200:
        filename = "./docs/" + pdf_link.split('/')[-1] + ".pdf"
        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded {filename}")
    else:
        print(f"Failed to download {pdf_link}")


# 提取链接
arxiv_links = find_arxiv_links(s)

# 使用ThreadPoolExecutor并行下载
with ThreadPoolExecutor(max_workers=10) as executor:
    executor.map(download_pdf, arxiv_links)
```
